```python
import json
import matplotlib
matplotlib.use("Agg")   # backend non interactif, pas de fenêtre, pas de crash
import matplotlib.pyplot as plt

# Chargement
with open("history_dagger.json", "r") as f:
    history = json.load(f)

# Création de la figure SANS l'afficher
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.plot(history["supervised_loss"], marker='o')
plt.title("Loss imitation")

plt.subplot(1,3,2)
plt.plot(history["reward_collect"], marker='o')
plt.title("Reward collect")

plt.subplot(1,3,3)
plt.plot(history["reward_eval"], marker='o')
plt.title("Reward evaluation")

plt.tight_layout()

# SAUVEGARDE uniquement, PAS d'affichage
plt.savefig("courbes_dagger.png", dpi=150)
plt.close()  # ferme proprement la figure pour libérer la mémoire

print("✓ Courbes sauvegardées dans 'courbes_dagger.png' (aucune ouverture de fenêtre)")

```
